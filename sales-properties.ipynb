# Properties with Pool
Since we have lists for the zip codes 92252 and 92284 indicating that properties for sale have a pool, we will proceed to create a variable called "pool" with a value of 1.

```python
# Create a column 'pool' with value 1 in 'zip_pool_92252' and in 'zip_pool_92284'

zip_pool_92252['pool'] = 1

zip_pool_92284['pool'] = 1
```

Once the variable is created, we proceed to select only the "pool" variable and the "Url". The aim is to add information about the availability of a pool in the total list of properties. Finally, we will have values of 1 if the property has a pool and 0 otherwise.







```python
# Select the column 'pool' and 'Url'

zip_pool_92252 = zip_pool_92252[['pool', 'Url']]
zip_pool_92284 = zip_pool_92284[['pool', 'Url']]

# Merge 'zip_92252' and 'zip_pool_92252' by 'Url'

zip_92252 = zip_92252.merge(zip_pool_92252, on='Url', how='left')

# Merge 'zip_92284' and 'zip_pool_92284' by 'Url'

zip_92284 = zip_92284.merge(zip_pool_92284, on='Url', how='left')

# Fill the NaN values of 'pool' with 0

zip_92252['pool'] = zip_92252['pool'].fillna(0)
zip_92284['pool'] = zip_92284['pool'].fillna(0)

# Convert 'pool' to int

zip_92252['pool'] = zip_92252['pool'].astype(int)
zip_92284['pool'] = zip_92284['pool'].astype(int)
```

## All properties
Since the zipcodes 92314 and 92315 do not contain properties with a pool, we directly assign them the value of 0 in the "pool" variable. This way, we can unify all the total lists into one: zip_df.

```python
# Create a column 'pool' with value 0 in 'zip_92314' and 'zip_92315'

zip_92314['pool'] = 0
zip_92315['pool'] = 0

# Concatenate 'zip_92252', 'zip_92284', 'zip_92314' and 'zip_92315'

zip_df = pd.concat([zip_92252, zip_92284, zip_92314, zip_92315], ignore_index=True)

# Display the first 5 rows of 'zip_df'

zip_df.head()
```

Since we have unnecessary variables, we will proceed to delete them, such as Days on Zillow, since it is not specified in any, Country, since we know they are all in the USA, sgapt, since we can obtain the same information in statusText, Lot Size since there is no value in it, isZillowOwned since none belong to Zillow, and finally statusType since they are all for sale.

```python
# Drop days on zillow, Country, sgapt, Lot Size, isZillowOwned and statusType columns.

zip_df = zip_df.drop(['Days On Zillow', 'Country', 'sgapt', 'Lot Size', 'statusType', 'isZillowOwned'], axis=1)

# Display the first 5 rows of 'zip_df'

zip_df.head()
```

```python
# Define the characteristics/features to visualize
features = ['Bedrooms', 'Bathrooms', 'Living Area', 'Home Type', 'pool']

# Plotting characteristics of each property
plt.figure(figsize=(12, 10))

# Loop through each feature and create subplots
for i, feature in enumerate(features, 1):
    plt.subplot(3, 2, i)
    if zip_df[feature].dtype == 'object':  # For categorical variables
        sns.countplot(x=feature, data=zip_df)
        plt.title(f'Distribution of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Count')
        
    else:  # For numerical variables
        sns.histplot(x=feature, data=zip_df, kde=True)
        plt.title(f'Distribution of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
